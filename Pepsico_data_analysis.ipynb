**Explore the dataset**

```python
# total products launched by pepsico

data['Product Name'].value_counts().sum()
```

```python
# in which category more and less products has been launched

mcl = data['Category'].value_counts().idxmax()
mcv = data['Category'].value_counts().max()
print(f"Pepsico has launched more products in {mcl} category ( {mcv} products )")

lcl = data['Category'].value_counts().idxmin()
lcv = data['Category'].value_counts().min()
print(f"Pepsico has launched less product in {lcl} category ( {lcv} products )")
```

```python
# how many active and discontinued products found

data['Status'].value_counts()
```

```python
# region served vs status

rs = data.groupby(['Region Served' , 'Status']).size().unstack(fill_value = 0)

rs
```

```python
plt.figure(figsize = (20,8))
rs.plot(kind = 'bar')
plt.title("Region Served vs Status")
plt.show()
```

```python
# products in each region

per = data['Region Served'].value_counts()

per
```

```python
plt.figure(figsize = (20,8))
plt.title("No of products in each region")
sns.countplot(data = data , x = 'Region Served')
plt.xticks(rotation = 45);
```

```python
# Trend of product launched per year
pyl = data.groupby('Year Launched').size().reset_index(name = 'Product Count')

plt.figure(figsize = (20,8))
sns.lineplot(data = pyl , x = "Year Launched" , y = "Product Count")
plt.title("Trend of product launched per year")
plt.show()
```

```python
# In which year pepsico has launched more and less products

mpy = data['Year Launched'].value_counts().idxmax()
mpv = data['Year Launched'].value_counts().max()

lpy = data['Year Launched'].value_counts().idxmin()
lpv = data['Year Launched'].value_counts().min()

print(f"Pepsico has launched more products in {mpy} year ( {mpv} products )")
print(f"Pepsico has launched less products in {lpy} year ( {lpv} products )")
```

```python
# what was pepsico's first launched product

fl = min(data['Year Launched'])
flp = data[data['Year Launched']==fl]
flp
```

```python
# What was the latest product launched by pepsico

ll = max(data['Year Launched'])
llp = data[data['Year Launched']==ll]
llp
```

```python
# No of products owned by each owner ( ownership )

plt.figure(figsize = (20,8))
plt.title("No of products vs Ownership")
sns.countplot(data = data , x = 'Ownership')
plt.xticks(rotation = 45);
```

```python
# trend of status over year

tsy = data.groupby(['Year Launched' , 'Status']).size().unstack(fill_value = 0)

plt.figure(figsize = (20,8))
sns.lineplot(tsy)
plt.title("Trend of status over year")
plt.show()
```

```python
ppc = data.groupby(['Category' , 'Region Served']).size()
ppc
```

```python
# Plotting the preference of product categories by region using Seaborn
plt.figure(figsize=(20, 8))
sns.countplot(data=data ,  x='Region Served', hue='Category')
plt.title('Preference of Product Categories by Region')
plt.xlabel('Region')
plt.ylabel('Number of Products')
plt.legend(title='Category')
plt.xticks(rotation=45);
plt.show()
```

```python
# category vs year

category_year = data.groupby(['Year Launched' , 'Category']).size().unstack(fill_value = 0)

category_year.plot(kind = 'line' , figsize = (20,8))
plt.title("Category trend over time")
plt.show()
```
